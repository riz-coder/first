# LINE PLOT

```python
#Importing the Libs
from numpy import median
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Load the Dataset
flower = sns.load_dataset('iris')
flower.head(5)

#Draw a line plot
sns.lineplot(x='sepal_length', y='sepal_width', data=flower)
plt.show()
```

```python
#Adding Title
sns.lineplot(x='sepal_length', y='sepal_width', data=flower)
plt.title('Iris Graph Repres')
plt.show()
```

```python
#Adding Limits
sns.lineplot(x='sepal_length', y='sepal_width', data=flower)
plt.title('Iris Graph Repres')
plt.xlim(4)       #X-axis Starts with 4
plt.ylim(2)
plt.show()
```

```python
#Set the Styles
sns.lineplot(x='sepal_length', y='sepal_width', data=flower)
plt.title('Iris Graph Repres')
sns.set_style(style=None, rc=None)
sns.set_style('dark')
plt.show()
```

```python
#Changing the Size Of Fig
sns.lineplot(x='sepal_length', y='sepal_width', data=flower)
plt.title('Iris Graph Repres')
plt.figure(figsize=(4,2));
plt.show()
```

# BARPLOT

```python
#Importing the Libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flower.head(5)
```

```python
#Draw a Bar plot
sns.barplot(x='species', y='sepal_width', data=flower)
plt.show()
```

```python
#Load the Another Dataset
kashti = sns.load_dataset('titanic')
kashti.head(5)
```

```python
#Draw a Bar plot
sns.barplot(x='who', y='alone', data=kashti, hue='sex')
plt.show()
```

```python
#Set the Order
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'])
plt.show()
```

```python
#Change the Colour 
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'], color='red')
plt.show()
```

```python
#Change the Pallete
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'], palette='dark')
plt.show()
```

```python
#Remove the Error Bars(ci)
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'], palette='pastel', ci=None)
plt.title('Bari Kashti')
plt.show()
```

```python
#Adding the Estimator instead of ci
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'], palette='pastel', estimator=median)
plt.title('Bari Kashti')
plt.show()
```

```python
#Changing the Color Saturation
sns.barplot(x='who', y='alone', data=kashti, hue='sex', order=['woman', 'man', 'child'], palette='pastel', saturation=6)
plt.title('Bari Kashti')
plt.show()
```

```python
#Graph Changed to Horizontal
sns.barplot(x='alone', y='who', data=kashti, hue='sex', order=['woman', 'man', 'child'], palette='pastel', saturation=6)
plt.title('Bari Kashti')
plt.show()
```

```python
#Customize the Grpah as your own
#Linewidth increases the Borders
#Errcolor changes the 'ci' of Plot, 0 is black
#Edgecolor changes the border of Plot, 0 is black
#Facecolor basically shows the main color of graph and might be work on RGB.
sns.barplot(x='who', y='alone',data=kashti,linewidth=6, errcolor='0.8', edgecolor='0', facecolor=(0.5,0.1,1,1))
plt.title('Bari Kashti')
plt.show()
```

# BOXPLOT

```python
#Plot the Boxplot
sns.set(style='whitegrid')
sns.boxplot(x='class', y='fare',data=kashti)
plt.title('Bari Kashti')
plt.show()
```

Its gives us Quartile info like median, 25%, 75% so on

```python
#Lets try with different dataset
tip = sns.load_dataset('tips')
tip.head(5)
```

```python
#Simple Box Plot
sns.boxplot(x='day', y='tip',data=tip)
plt.title('Naseeb Biryani')
plt.show()
```

```python
#Let's play with Pandas on Tip DS
tip.describe()      #Give the idea of numeric data in dataset
```

Numeric Values usually plot on x-axis while Categorical on y-axis

```python
#If you know the Graphical rep of only one Col
sns.boxplot(x=tip['total_bill']);
```

```python
#Simple Box Plot with Hori touch while Dodge allow us to separate the Hue while its True
sns.boxplot(x='tip', y='day',hue='smoker',data=tip, palette='Set2', dodge=True)
plt.title('Naseeb Biryani')
plt.show()
```

```python
#Change color with Hex number
sns.boxplot(x='tip', y='day',hue='smoker',data=tip, color='#05470f')
plt.title('Naseeb Biryani')
plt.show()
```

```python
#Change the HUE color(Basically we change the hue color with palette option)
sns.boxplot(x='tip', y='day',hue='smoker',data=tip, palette='Set1')
plt.title('Naseeb Biryani')
plt.show()
```
